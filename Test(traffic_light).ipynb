```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Function to detect traffic light colors
def detect_traffic_light(image_path):
    # Load the image
    image = cv2.imread(image_path)
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)  # Convert to RGB for display
    hsv = cv2.cvtColor(image, cv2.COLOR_RGB2HSV)   # Convert to HSV

    # Define color ranges for Red, Yellow, and Green in HSV
    lower_red1, upper_red1 = np.array([0, 100, 100]), np.array([10, 255, 255])
    lower_red2, upper_red2 = np.array([160, 100, 100]), np.array([180, 255, 255])  # Second range for red
    lower_yellow, upper_yellow = np.array([15, 100, 100]), np.array([35, 255, 255])
    lower_green, upper_green = np.array([40, 50, 50]), np.array([90, 255, 255])

    # Create masks for colors
    mask_red = cv2.inRange(hsv, lower_red1, upper_red1) | cv2.inRange(hsv, lower_red2, upper_red2)
    mask_yellow = cv2.inRange(hsv, lower_yellow, upper_yellow)
    mask_green = cv2.inRange(hsv, lower_green, upper_green)

    # Find contours for each mask
    def find_light(mask, color_name):
        contours, _ = cv2.findContours(mask, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
        for cnt in contours:
            if cv2.contourArea(cnt) > 50:  # Filter small areas
                x, y, w, h = cv2.boundingRect(cnt)
                cv2.rectangle(image, (x, y), (x + w, y + h), (255, 255, 255), 2)
                cv2.putText(image, color_name, (x, y - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 255, 255), 2)

    find_light(mask_red, "RED")
    find_light(mask_yellow, "YELLOW")
    find_light(mask_green, "GREEN")

    # Show the image
    plt.figure(figsize=(6, 6))
    plt.imshow(image)
    plt.axis("off")
    plt.show()

# Upload image in Colab
from google.colab import files
uploaded = files.upload()

# Get the uploaded image path and run detection
image_path = list(uploaded.keys())[0]  # Get the first uploaded image
detect_traffic_light(image_path)
```

```python
from google.colab import files
uploaded = files.upload()
image_path = list(uploaded.keys())[0]  # Get uploaded file name


results = model(image_path)
detections = results.xyxy[0].cpu().numpy()
print(detections)
```

```python
confidence_threshold = 0.5 # confidence interval for traffic light detection
import numpy as np
# filter detections for class_id == 2 (traffic lights) and confidence > threshold
traffic_lights = []
for detection in detections:
  xmin, ymin, xmax, ymax, confidence, class_id = detection
  if class_id == 9 and confidence > confidence_threshold:
    traffic_lights.append([xmin, ymin, xmax, ymax, confidence])

traffic_lights = np.array(traffic_lights)
print("Filtered Traffic Lights:", traffic_lights)
```

```python
from PIL import Image, ImageDraw
import os
import matplotlib.pyplot as plt

img = Image.open(image_path)

# calling 'img 59666.png' as img
image_with_boxes = img.copy()
draw = ImageDraw.Draw(image_with_boxes)

#Draw a bounding box around each traffic light
for x_min, y_min, x_max, y_max, confidence in traffic_lights:
  draw.rectangle([x_min, y_min, x_max, y_max], outline="red", width=3)
  draw.text((x_min, y_min), f"{confidence:.2f}", fill="white")

# Display the image
plt.figure(figsize=(10, 10))
plt.imshow(image_with_boxes)
plt.axis('off')
plt.show()

display(image_with_boxes)
print(f"Loaded image: {image_with_boxes}")
```
